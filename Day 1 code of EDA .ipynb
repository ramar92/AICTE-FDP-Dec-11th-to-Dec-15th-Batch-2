```python
# IQR

Q1 = np.percentile(df2['sepal_width'], 25, interpolation='midpoint') 

Q3 = np.percentile(df2['sepal_width'], 75, interpolation='midpoint') 

IQR =  Q3 - Q1 

print(IQR)
print('old shape', df.shape)
```

```python
print('old shape', df2.shape)
```

```python
upper = np.where(df2['sepal_width'] >= (Q3-1.5*IQR))
lower = np.where(df2['sepal_width'] <= (Q1-1.5*IQR))
```

```python
df2.drop(upper[0], inplace=True)
df2.drop(lower[0], inplace=True)
```
